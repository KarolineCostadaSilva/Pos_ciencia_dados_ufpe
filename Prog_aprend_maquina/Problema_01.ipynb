Gerar uma matriz, a partir de um ou de ambos os textos anexo.

Temos que ser capazes de voltar ao texto a partir da matriz. Podemos abrir mão de caracteres especiais e pontuação. Deve ser entregue a matriz, um texto explicando a estratégia e o código fonte utilizado para gerar a matriz (quando houver).
-------

### Com ASCII

```python
def text_to_matrix(text):
    # Limpa o texto, removendo caracteres especiais e pontuações
    cleaned_text = ''.join(e for e in text if e.isalnum() or e.isspace()).split()

    # Encontra o comprimento máximo da palavra no texto
    max_word_length = max([len(word) for word in cleaned_text])

    # Inicializa a matriz
    matrix = []

    # Converte cada palavra em uma lista de valores ASCII
    for word in cleaned_text:
        ascii_values = [ord(char) for char in word]
        # Preenche a palavra com zeros até o comprimento máximo
        while len(ascii_values) < max_word_length:
            ascii_values.append(0)
        matrix.append(ascii_values)

    return matrix

def matrix_to_text(matrix):
    # Converte a matriz de volta ao texto
    words = []
    for row in matrix:
        word = ''.join([chr(value) for value in row if value != 0])
        words.append(word)
    return ' '.join(words)

# Leitura do arquivo
with open(base_text_1, "r") as file1:
    file_content1 = file1.read()
    print("Conteúdo Original do Arquivo:")
    print(file_content1)

    matrix = text_to_matrix(file_content1)
    reconstructed_text = matrix_to_text(matrix)

    print("\nMatriz:")
    print(matrix)
    print("\nTexto Reconstruído:")
    print(reconstructed_text)
```

### Com Word2vec

O Word2Vec é uma técnica popular para converter palavras em vetores densos que capturam semântica e contexto. Cada palavra é representada como um vetor de n dimensões. Ao utilizar o Word2Vec, cria-se uma matriz onde cada linha corresponde a uma palavra do texto e cada coluna é uma das dimensões do vetor Word2Vec daquela palavra.

```python
from gensim.models import Word2Vec
import numpy as np

# O texto em uma lista
sentences = [file_content1.lower().split()]

# Treinar um modelo Word2Vec
model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)
model.train(sentences, total_examples=model.corpus_count, epochs=10)

def text_to_matrix_w2v(text, model):
    words = text.lower().split()
    matrix = []
    for word in words:
        if word in model.wv:
            matrix.append(model.wv[word])
        else:
            matrix.append(np.zeros(model.vector_size))  # vetor nulo para palavras desconhecidas
    return np.array(matrix)

# Teste
matrix_w2v = text_to_matrix_w2v(file_content1, model)
print(matrix_w2v)
```

### Combinação ASCII + Word2Vec

```python
# 1. Treinar um modelo Word2Vec
sentences = [file_content1.lower().split()]
model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)
model.train(sentences, total_examples=model.corpus_count, epochs=10)

def get_cleaned_words(text):
    return ''.join(e for e in text if e.isalnum() or e.isspace()).split()

# 2. Função para converter texto para matriz usando codificação ASCII
def text_to_ascii_matrix(text):
    cleaned_words = get_cleaned_words(text)
    max_word_length = max([len(word) for word in cleaned_words])
    matrix = []
    for word in cleaned_words:
        ascii_values = [ord(char) for char in word]
        while len(ascii_values) < max_word_length:
            ascii_values.append(0)
        matrix.append(ascii_values)
    return np.array(matrix)

# 3. Função para converter texto para matriz usando Word2Vec
def text_to_w2v_matrix(text, model):
    cleaned_words = get_cleaned_words(text)
    matrix = []
    for word in cleaned_words:
        if word in model.wv:
            matrix.append(model.wv[word])
        else:
            matrix.append(np.zeros(model.vector_size))
    return np.array(matrix)

# 4. Função combinada
def combined_text_to_matrix(text, model):
    ascii_matrix = text_to_ascii_matrix(text)
    w2v_matrix = text_to_w2v_matrix(text, model)

    # Combinando as duas matrizes horizontalmente
    return np.hstack((ascii_matrix, w2v_matrix))

# Teste
combined_matrix = combined_text_to_matrix(file_content1, model)
print(combined_matrix)
```
